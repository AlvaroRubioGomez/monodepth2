```python
from skimage.color import rgb2gray
from skimage import data
from skimage.segmentation import active_contour
from skimage.draw import polygon_perimeter

from skimage.filters import sobel
from skimage.feature import canny
from skimage import measure
from skimage.io import imread

```

## YOLO algorithm: 
To establish the bounding box of the person

```python
from utils import *
from darknet import Darknet
```

```python
img = imread("H_CB_07_8_disp.jpeg")
img_gray = rgb2gray(img)

#mask
mask = np.ones(img_gray.shape)
#yolo bounding box parameters
for x in range(500,900): 
    for y in range(200,1080):
        mask[y][x]= 0;
mask = mask < 1

#Edge filters
img_edges_sobel = sobel(img_gray, mask=None) #sobel
img_edges_canny = canny(img_gray, sigma=2.0, low_threshold=0.1, high_threshold=0.3, mask=mask) #canny

#Contours detectors
level = 0.05
contours_sobel = measure.find_contours(img_edges_sobel, level,'high','high')
contours_canny = measure.find_contours(img_edges_canny, level, 'high','high')

#plot sobel filter
fig1, ax1 = plt.subplots()
ax1.imshow(img_edges_sobel, cmap=plt.cm.gray)

#plot canny filter
fig2, ax2 = plt.subplots()
ax2.imshow(img_edges_canny, cmap=plt.cm.gray)

if(True):
    for n, contour in enumerate(contours_sobel):    
        ax1.plot(contour[:, 1], contour[:, 0], linewidth=2)
    for n, contour in enumerate(contours_canny): 
        ax2.plot(contour[:, 1], contour[:, 0], linewidth=2)    
    
plt.show()


```

```python
img = data.imread("H_CB_07_8_disp.jpeg")
img = rgb2gray(img)

#img shape
img_h = img.shape[0]
img_w = img.shape[1]
print(img_w, ",", img_h)

#Define rectangle
rr, cc = polygon_perimeter([1080,200,200,1080],[500,500,900,900],shape=img.shape, clip=True)
init = np.array([rr, cc]).T


snake = active_contour(img,init, alpha=0.001, beta=0.001, w_line=10, w_edge=5, 
                       gamma=0.01, max_px_move=5.0, max_iterations=200)
                       

fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(img, cmap=plt.cm.gray)
ax.plot(init[:, 1], init[:, 0], '--r', lw=3)
ax.plot(snake[:, 1], snake[:, 0], '-b', lw=3)
ax.set_xticks([]), ax.set_yticks([])
ax.axis([0, img.shape[1], img.shape[0], 0])

plt.show()
```
